```python
import matplotlib.pyplot as plt
import seaborn as sns
import pandas as pd

df = pd.read_csv('shopping_trends.csv')
# print(df.to_string())

sns.countplot(
    x="Payment Method",
    data=df,
    order=df["Payment Method"].value_counts().index
)

plt.title("Most Used Payment Method")
plt.xlabel("Payment Method")
plt.ylabel("Count")
plt.xticks(rotation=45)

plt.show()
```

```python
df1 = df.head(10)

sns.barplot(
    data=df1,
    x="Location",
    y="Purchase Amount (USD)",
    palette="viridis",
    hue='Location'
)

plt.title("Amount Generated by Locations")
plt.xlabel("Location")
plt.ylabel("Count")
plt.xticks(rotation=90)

plt.show()
```

```python
sns.histplot(data=df, x="Age", kde=True)

plt.title("Distribution of Age")
plt.xlabel("Age")
plt.ylabel("Frequency")

plt.show()
```

```python
sns.countplot(data=df, x="Color", hue="Gender")

plt.title("Quantity of Purchases by Color and Gender")
plt.xlabel("Color")
plt.ylabel("Count")
plt.xticks(rotation=45)

plt.show()
```
